```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('eda.csv')
```

```python
df.head()
```

```python
df.shape
```

### Checking the count of cars in the in Car inventory

```python
plt.figure(figsize = (20, 15))
sns.countplot(y = df.Name)
plt.title("Car companies with their cars", fontsize = 20)
plt.show()
```

```python
status_df = df.loc[df['Status'] != "unknown"]
```

```python
status_df.shape
```

### Comparing the price in terms of rating

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(x = df.Rating, y = df.Price)
plt.title("Car companies with their cars", fontsize = 20)
plt.show()
```

```python
Cars = list(df.Name.value_counts().index)
counts = list(df.Name.value_counts().values)

fig, ax = plt.subplots(figsize=(16, 13))

# Plot horizontal lines
plt.hlines(Cars[::-1], linewidths=5, xmin=0, xmax=counts[::-1])

# Plot circles
plt.scatter(counts[::-1], Cars[::-1], color='red', alpha=1, s=100)

# sns.barplot(x=counts, y=departments)
plt.title('Distribution of Cars by Brands', fontsize=20, fontweight='bold')
plt.xlabel('Count', fontsize=17)
plt.ylabel('Cars', fontsize=17)
plt.yticks(size=15)
plt.grid()
plt.xticks(range(0, 1200, 200), size=14);
plt.show()
```

### Average price of cars based on the Car Brand

```python
fig, ax = plt.subplots(figsize=(95, 50))
plt.rcdefaults()
plt.rcParams.update({'font.size':45})
sns.barplot(df['Name'],df['Price'])
ax.set_xlabel("Brand Name")
ax.set_ylabel('Average prices')
ax.set_title('Distribution of Cars prices based on the Car Brand')
#ax.set_xticklabels(avg_price, rotation=30)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
plt.grid()
plt.tight_layout()
plt.show();
```

```python
toyota_df = df[(df['Name'] == 'Toyota') & (df['Year'] != 2022) & (df['Year'] != 2023)]
```

```python
toyota_df
```

### Looking at the average price of cars based on the Car Class

```python
fig, ax = plt.subplots(figsize=(95, 50))
plt.rcdefaults()
plt.rcParams.update({'font.size':45})
sns.barplot(toyota_df['Class'],toyota_df['Price'])
ax.set_xlabel("Car Class")
ax.set_ylabel('Average prices')
ax.set_title('Distribution of Cars prices based on the Car Class')
#ax.set_xticklabels(avg_price, rotation=30)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
plt.grid()
plt.tight_layout()
plt.show();
```

```python
avg_price_year = toyota_df.groupby("Year")["Price"].mean().sort_values(ascending= False)
```
