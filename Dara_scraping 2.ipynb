```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
link1 = []
```

```python
for i in range(1,2000):
    website = "https://www.cars.com/shopping/results/?page=" + str(i) + "&page_size=20&list_price_max=&makes[]=&maximum_distance=all&models[]=&stock_type=used&zip="
       
    #request to the website
    response = requests.get(website)
    
    #creating the soup object
    soup = BeautifulSoup(response.content, "html.parser")
    
    #getting all the page results in one variable
    #results = soup.find_all('div', {'class':'vehicle-card'})
    for link in soup.find_all('a', 
                          attrs={'href': re.compile("^/vehicledetail/")}):
    # display the actual urls
        a = link.get('href')
        link1.append(a)
        
```

```python
name = []
title = []
mileage = []
price = []
deal_status = []
rating = []
rating_count = []
features = []
```

```python
for container in containers.values():
    websites = container
    
    
    for website in websites:
    #request to the website
        response = requests.get(website)
    
    #creating the soup object
        soup = BeautifulSoup(response.content, "html.parser")
    
    #getting all the page results in one variable
    #results = soup.find_all('section', {'class':'listing-overview'})
    
    #for result in results:
        #name
        try:
            a = soup.find('p',{'class':'new-used'}).get_text()
            name.append(a)
        except:
            name.append('Nan')
        
        try:
            a = soup.find('h1',{'class':'listing-title'}).get_text()
            title.append(a)
        except:
            title.append('Nan')
        
        try:
            a = soup.find('div',{'class':'listing-mileage'}).get_text()
            mileage.append(a)
        except:
            mileage.append('Nan')
        
        try:
            a = soup.find('span',{'class':'primary-price'}).get_text()
            price.append(a)
        except:
            price.append('Nan')
        
        try:
            a = soup.find('span',{'class':'sds-badge__label'}).get_text()
            deal_status.append(a)
        except:
            deal_status.append('Nan')
    
        try:
            a = soup.find('div',{'class':'auto-corrected-feature-list'}).get_text()
            features.append(a)
        except:
            features.append('Nan')
    
        try:
            a = soup.find('span',{'class':'sds-rating__count'}).get_text()
            rating.append(a)
        except:
            rating.append('Nan')
        

    
        try:
            a = soup.find('a',{'class':'sds-rating__link sds-button-link'}).get_text()
            rating_count.append(a)
        except:
            rating_count.append('Nan')
        
```
